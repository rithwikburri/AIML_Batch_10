```python
# Cell 1
!pip install --quiet plotly==5.15.0 kaleido scikit-learn ipywidgets

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from IPython.display import display
import os

pd.set_option('display.max_columns', 200)
print("Libraries loaded.")
```

```python
# Cell 2
path = "/content/IPL dataset final.csv"
df = pd.read_csv(path, low_memory=False)
print("Loaded:", path)
print("Shape:", df.shape)
display(df.head())
print("\nColumns:\n", df.columns.tolist())
```

```python
# Cell 3
# Confirm required columns exist (based on your provided list)
required = ['Player','TEAM','Runs','BF','4s','6s','SR','TRuns','TBF','TSR','B_Wkts','B_Econ','B_Balls']
missing = [c for c in required if c not in df.columns]
print("Missing required columns (if any):", missing)
# Show sample of relevant columns
sample_cols = ['Player','COUNTRY','TEAM','Paying_Role','Runs','BF','SR','4s','6s','TRuns','TBF','TSR','B_Wkts','B_Econ','B_Balls','SOLD_PRICE']
present = [c for c in sample_cols if c in df.columns]
display(df[present].head())
```

```python
# Cell 4
df = df.copy()

# canonical mapping using the columns you have
player_col = 'Player'
team_col = 'TEAM'

# Batting (single-season and career totals available)
# Single-season fields (may be per dataset)
if 'Runs' in df.columns:        df['runs_total'] = pd.to_numeric(df['Runs'], errors='coerce').fillna(0)
if 'BF' in df.columns:          df['balls_faced'] = pd.to_numeric(df['BF'], errors='coerce').fillna(0)
if '4s' in df.columns:          df['fours'] = pd.to_numeric(df['4s'], errors='coerce').fillna(0)
if '6s' in df.columns:          df['sixes'] = pd.to_numeric(df['6s'], errors='coerce').fillna(0)
if 'SR' in df.columns:         df['strike_rate'] = pd.to_numeric(df['SR'], errors='coerce').fillna(0)

# Career totals (prefixed T*)
if 'TRuns' in df.columns:       df['runs_total_career'] = pd.to_numeric(df['TRuns'], errors='coerce').fillna(0)
if 'TBF' in df.columns:         df['balls_faced_career'] = pd.to_numeric(df['TBF'], errors='coerce').fillna(0)
if 'TSR' in df.columns:         df['strike_rate_career'] = pd.to_numeric(df['TSR'], errors='coerce').fillna(0)

# Bowling
if 'B_Wkts' in df.columns:      df['wickets'] = pd.to_numeric(df['B_Wkts'], errors='coerce').fillna(0)
if 'B_Econ' in df.columns:      df['economy'] = pd.to_numeric(df['B_Econ'], errors='coerce')
else:                           df['economy'] = np.nan
if 'B_Balls' in df.columns:     df['balls_bowled'] = pd.to_numeric(df['B_Balls'], errors='coerce').fillna(0)

# Sold price: keep as string for now
if 'SOLD_PRICE' in df.columns:
    df['sold_price'] = df['SOLD_PRICE'].astype(str)

# Some derived safe columns
df['boundary_runs'] = df[['fours','sixes']].fillna(0)['fours']*4 + df[['fours','sixes']].fillna(0)['sixes']*6
df['matches'] = pd.to_numeric(df['Mat'], errors='coerce').fillna(0)

print("Canonical columns created. Example:")
display(df[[player_col, team_col, 'runs_total','balls_faced','fours','sixes','strike_rate','runs_total_career','wickets','economy','sold_price']].head())
```

```python
# Cell 5
players = df[[player_col, team_col, 'runs_total','balls_faced','fours','sixes','strike_rate',
              'runs_total_career','balls_faced_career','strike_rate_career',
              'wickets','economy','balls_bowled','matches','sold_price']].copy()

players.columns = ['Player','Team','Runs','Balls','4s','6s','SR',
                   'Runs_Career','Balls_Career','SR_Career',
                   'Wickets','Economy','BallsBowled','Matches','SoldPrice']

# Derived metrics (avoid division by zero)
players['StrikeRate'] = (players['Runs'] / players['Balls'].replace({0:np.nan}) * 100).round(2).fillna(0)
players['BoundaryPercent'] = ((players['4s']*4 + players['6s']*6) / players['Balls'].replace({0:np.nan}) * 100).round(2).fillna(0)

# Order cols
display(players.sort_values('Runs', ascending=False).head(20))
```

```python
# Cell 6
top_n = 15
top_batsmen = players.sort_values('Runs', ascending=False).head(top_n)

fig = px.bar(top_batsmen, x='Runs', y='Player', orientation='h', color='Team',
             title=f"Top {top_n} Batsmen by Runs",
             hover_data=['StrikeRate','4s','6s','Matches'])
fig.update_layout(height=600, yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
# Cell 7
if players['Wickets'].sum() == 0:
    print("No bowling wickets data available in this dataset.")
else:
    top_bowlers = players.sort_values('Wickets', ascending=False).head(15)
    fig = px.bar(top_bowlers, x='Wickets', y='Player', orientation='h', color='Team',
                 title="Top 15 Bowlers by Wickets", hover_data=['Economy','BallsBowled'])
    fig.update_layout(height=600, yaxis={'categoryorder':'total ascending'})
    fig.show()
```

```python
# Cell 8
def radar_profile(name):
    if name not in players['Player'].values:
        print("Player not found. Use exact name from players['Player'].head()")
        return
    p = players[players['Player']==name].iloc[0]
    # choose features and normalize relative to 95th percentile for nice scale
    caps = {}
    caps['Runs'] = min(100, p['Runs'] / players['Runs'].quantile(0.95) * 100)
    caps['SR']   = min(100, p['StrikeRate'] / players['StrikeRate'].quantile(0.95) * 100)
    caps['Bound%'] = min(100, p['BoundaryPercent'] / players['BoundaryPercent'].quantile(0.95) * 100)
    caps['Wkts'] = min(100, p['Wickets'] / players['Wickets'].quantile(0.95) * 100) if players['Wickets'].max()>0 else 0
    caps['Econ'] = 100 - min(100, (p['Economy'] / players['Economy'].quantile(0.95) * 100)) if players['Economy'].notna().sum()>0 else 0

    cats = list(caps.keys())
    vals = list(caps.values())
    vals += vals[:1]; cats += cats[:1]
    fig = go.Figure(data=go.Scatterpolar(r=vals, theta=cats, fill='toself'))
    fig.update_layout(title=f"Player Radar — {name}", polar=dict(radialaxis=dict(visible=True, range=[0,100])))
    fig.show()

# Example: replace name with an actual value from players['Player'].head()
radar_profile(players['Player'].iloc[18])
```

```python
# Cell 9
numeric_cols = ['Runs','Balls','SR','4s','6s','BoundaryPercent','Wickets','Economy','Matches']
present = [c for c in ['Runs','Balls','SR','4s','6s','BoundaryPercent','Wickets','Economy','Matches'] if c in players.columns]
corr = players[present].corr().round(2)
fig = px.imshow(corr, text_auto=True, aspect='auto', title="Correlation Heatmap (Player Metrics)")
fig.update_layout(height=600)
fig.show()
```

```python
# Cell 10
cluster_cols = [c for c in ['Runs','SR','BoundaryPercent','Wickets','Economy'] if c in players.columns]
if len(cluster_cols) < 2:
    print("Not enough numeric columns for clustering. Found:", cluster_cols)
else:
    X = players[cluster_cols].fillna(0)
    Xs = StandardScaler().fit_transform(X)
    kmeans = KMeans(n_clusters=4, random_state=42).fit(Xs)
    players['Cluster'] = kmeans.labels_.astype(str)
    pca = PCA(n_components=2).fit_transform(Xs)
    players['pca1'] = pca[:,0]; players['pca2'] = pca[:,1]

    fig = px.scatter(players, x='pca1', y='pca2', color='Cluster', hover_name='Player',
                     title='Player Clusters (PCA projection)', width=900, height=600)
    fig.show()
    display(players[['Player','Team'] + cluster_cols + ['Cluster']].sort_values('Cluster').head(40))
```

```python
# Cell 11
ps = players.copy()

# Normalize selected metrics 0..1
metrics = []
if 'Runs' in ps.columns: metrics.append('Runs')
if 'SR' in ps.columns: metrics.append('SR')
if 'Wickets' in ps.columns: metrics.append('Wickets')
if 'Economy' in ps.columns: metrics.append('Economy')

for m in metrics:
    if m == 'Economy':
        ps['norm_' + m] = 1 - (ps[m] - ps[m].min()) / (ps[m].max() - ps[m].min() + 1e-9)
    else:
        ps['norm_' + m] = (ps[m] - ps[m].min()) / (ps[m].max() - ps[m].min() + 1e-9)

# example weights (tune as desired)
weights = {'Runs':0.45, 'SR':0.2, 'Wickets':0.25, 'Economy':0.10}
ps['ImpactScore'] = 0
for m in metrics:
    ps['ImpactScore'] += ps['norm_' + m] * weights.get(m, 0)

top_impact = ps.sort_values('ImpactScore', ascending=False).head(30)
display(top_impact[['Player','Team','ImpactScore'] + metrics].round(3))

fig = px.bar(top_impact.head(15), x='ImpactScore', y='Player', orientation='h', color='Team',
             title='Top 15 Players by Impact Score', hover_data=metrics)
fig.update_layout(height=600, yaxis={'categoryorder':'total ascending'})
fig.show()
```
